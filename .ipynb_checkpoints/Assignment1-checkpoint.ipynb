```python
#Question 1

import psycopg2
import pandas as pd

conn = psycopg2.connect(
    database="dvdrental",
    user="postgres",
    password="password",
    host="127.0.0.1",
    port="5432"
)

# SQL Query
sql = """
SELECT 
    DATE(payment_date) AS payment_day,   
    co.country,
    SUM(p.amount) AS daily_total_revenue
FROM payment p
JOIN customer cu ON p.customer_id = cu.customer_id
JOIN address a ON cu.address_id = a.address_id
JOIN city ci ON a.city_id = ci.city_id
JOIN country co ON ci.country_id = co.country_id
WHERE co.country IN ('United States', 'Canada', 'Mexico')
GROUP BY payment_day, co.country
ORDER BY payment_day, co.country;
"""

# Use pandas to read the query result into a DataFrame
df = pd.read_sql_query(sql, conn)


```

```python
import matplotlib.pyplot as plt

# Filter data for each country
us_data = df[df['country'] == 'United States']
canada_data = df[df['country'] == 'Canada']
mexico_data = df[df['country'] == 'Mexico']

# Plotting
plt.figure(figsize=(14,7))
plt.plot(us_data['payment_day'], us_data['daily_total_revenue'], label='United States')
plt.plot(canada_data['payment_day'], canada_data['daily_total_revenue'], label='Canada')
plt.plot(mexico_data['payment_day'], mexico_data['daily_total_revenue'], label='Mexico')

# Add title and labels
plt.title('Daily Total Revenue for Three Countries')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()

# Show the plot
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#Question 2

# Query to get film descriptions
sql = "SELECT film_id, title, description FROM film;"

# Use pandas to read the query result into a DataFrame
df = pd.read_sql_query(sql, conn)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

def get_sentiment(description):
    score = sia.polarity_scores(description)
    
    # Classify sentiment based on the compound score
    if score['compound'] >= 0.05:
        return 'positive'
    elif score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment'] = df['description'].apply(get_sentiment)
```

```python
# Visualize the distribution of sentiments
import matplotlib.pyplot as plt

# Plot the distribution of sentiments
df['sentiment'].value_counts().plot(kind='bar')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Question 3

sql = """
SELECT 
    co.country,
    SUM(p.amount) AS total_revenue
FROM payment p
JOIN customer cu ON p.customer_id = cu.customer_id
JOIN address a ON cu.address_id = a.address_id
JOIN city ci ON a.city_id = ci.city_id
JOIN country co ON ci.country_id = co.country_id
GROUP BY co.country
ORDER BY total_revenue DESC;
"""

# Use pandas to read the query result into a DataFrame
df_revenue = pd.read_sql_query(sql, conn)
```

```python
import matplotlib.pyplot as plt

# Adjust the figure size for more space between y-axis labels
plt.figure(figsize=(9, len(df_revenue) * 0.3))

# Bar chart
plt.barh(df_revenue['country'], df_revenue['total_revenue'], color='orange')

# Set title and labels
plt.title('Total Revenue by Country for the DVD Store')
plt.xlabel('Total Revenue')
plt.ylabel('Country')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
#Question 4

sql = """
SELECT 
    a.actor_id, 
    a.first_name || ' ' || a.last_name AS actor_name,
    SUM(p.amount) AS total_revenue
FROM actor a
JOIN film_actor fa ON a.actor_id = fa.actor_id
JOIN film f ON fa.film_id = f.film_id
JOIN inventory i ON f.film_id = i.film_id
JOIN rental r ON i.inventory_id = r.inventory_id
JOIN payment p ON r.rental_id = p.rental_id
GROUP BY a.actor_id
ORDER BY total_revenue DESC
LIMIT 10;
"""
df_actors_revenue = pd.read_sql_query(sql, conn)
```

```python
print(df_actors_revenue)
```

```python
# Load relevant tables into DataFrames
df_actor = pd.read_sql_query("SELECT * FROM actor", conn)
df_film_actor = pd.read_sql_query("SELECT * FROM film_actor", conn)
df_film = pd.read_sql_query("SELECT * FROM film", conn)
df_inventory = pd.read_sql_query("SELECT * FROM inventory", conn)
df_rental = pd.read_sql_query("SELECT * FROM rental", conn)
df_payment = pd.read_sql_query("SELECT * FROM payment", conn)

# Start by merging actor and film_actor
df_merged = df_actor.merge(df_film_actor, on='actor_id')

# Merge with film
df_merged = df_merged.merge(df_film, on='film_id', suffixes=('', '_from_film'))

# Merge with inventory
df_merged = df_merged.merge(df_inventory, on='film_id', suffixes=('', '_from_inventory'))

# Merge with rental using inventory_id
df_merged = df_merged.merge(df_rental, on='inventory_id', suffixes=('', '_from_rental'))

# Merge with payment using rental_id
df_merged = df_merged.merge(df_payment, on='rental_id', suffixes=('', '_from_payment'))

# Group by actor and aggregate by revenue
df_actor_revenue = df_merged.groupby(['actor_id', 'first_name', 'last_name'])['amount'].sum().reset_index()

# Rename columns for clarity
df_actor_revenue.columns = ['actor_id', 'first_name', 'last_name', 'total_revenue']

# Sort and Select Top 10
df_top_actors = df_actor_revenue.sort_values(by='total_revenue', ascending=False).head(10)

print(df_top_actors)
```

```python
#Question 5&6

import pytz
import pycountry

# Extract unique countries from the database
df_countries = pd.read_sql_query("SELECT DISTINCT country FROM country", conn)

# Function to convert country name to ISO country code
def get_country_code(country_name):
    try:
        return pycountry.countries.get(name=country_name).alpha_2
    except AttributeError:
        return None

# Map the countries to their ISO country codes
df_countries['Country_Code'] = df_countries['country'].apply(get_country_code)

# Function to fetch timezone using the country code
def get_timezone(country_code):
    if country_code:
        timezones = pytz.country_timezones.get(country_code)
        if timezones:
            return timezones[0]  # Return the first timezone if there are multiple
    return None

# Map the country codes to timezones
df_countries['Time_Zone'] = df_countries['Country_Code'].apply(get_timezone)

# Display the combined DataFrame
print(df_countries[['country', 'Country_Code', 'Time_Zone']])
```
